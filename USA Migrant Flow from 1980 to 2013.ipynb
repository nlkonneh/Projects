```python
import numpy as np
import pandas as pd
```

```python
df_usa = pd.read_excel('United States of America.xlsx')
df_usa
```

```python
print(df_usa.shape)
```

Clean up data. We will make some modifications to the original dataset to make it easier to create our visualizations. Refer to `Introduction to Matplotlib and Line Plots` lab for the rationale and detailed description of the changes.

```python
df_usa.head()
```

#### 1. Clean up the dataset to remove columns that are not informative to us for visualization (eg. Type, AREA, REG).

```python
df_usa.drop(['AREA', 'REG', 'DEV', 'Type', 'Coverage'], axis=1, inplace=True)
df_usa.head()
```

#### 2. Rename some of the columns so that they make sense.

```python
df_usa.rename(columns = {'OdName':'Country', 'AreaName':'Continent', 'RegName':'Region'}, inplace=True)

df_usa.head()
```

```python
df_usa.applymap(np.isreal).head() # this code finds any non-integer and returns a boolean 
```

#### 3. For consistency, ensure that all column labels of type string.

```python
df_usa.columns = list(map(str, df_usa.columns))
all(isinstance(column, str) for column in df_usa.columns)
```

#### 4. Set the country name as index - useful for quickly looking up countries using .loc method.

```python
df_usa.set_index('Country', inplace=True)
df_usa.head()
```

Notice how the country names now serve as **indices**.

#### 5. Add total column.

```python
df_usa['Total'] = df_usa.sum(axis=1)
df_usa.head()
```

```python
df_usa.drop(df_usa.index[[0]], inplace=True)
df_usa.head()
```

```python
print('Data dimension: ', df_usa.shape)
```

#### The shape above shows 221 rows and 38 columns including the newly added 'Total' column 

#### 6. Finally, let's create a list of years from 1980 - 2013. This will come in handy when we start plotting the data

```python
years = list(map(str, range(1980, 2014)))
years
```

# Visualizing Data using Matplotlib<a id="4"></a>

```python
%matplotlib inline

import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.style.use('ggplot')

print('Matplotlib ' + ' version: ', mpl.__version__)   # to check the latest version 
```

# Area Plots<a id="6"></a>

Top 5 countries that contributed immigrants to the US from 1980 to 2013

```python
df_usa.sort_values(['Total'], ascending=False, axis=0, inplace=True)

df_top5 = df_usa.head()

df_top5 = df_top5[years].transpose()

df_top5.head()
```

#### 7. Using the scripting layer (procedural method using matplotlib.pyplot as plt) to plot 

```python
df_top5.index = df_top5.index.map(int)
df_top5.plot(kind='area', stacked=False, figsize=(20, 10))

plt.title('Immigration Trend of top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.show()
```

The unstacked plot has a default transparency (alpha value) at 0.5. We can modify this value by passing in the `alpha` parameter.

```python
df_top5.index = df_top5.index.map(int)
df_top5.plot(kind='area', alpha=0.25, stacked=False, figsize=(20, 10))

plt.title('Immigration Trend of top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.show()
```

#### 7. Plotting using the artist layer (object oriented method using an Axes instance from Matplotlib

```python
ax = df_top5.plot(kind='area', alpha=0.35, figsize=(20, 10))

ax.set_title('Immigration Trend of Top 5 Countries')
ax.set_ylabel('Number of Immigrants')
ax.set_xlabel('Years')
```

#### 8. Countries that contributed the least to immigration to the United States from 1980 to 2013

```python
df_least5 = df_usa.tail(5)

df_least5 = df_least5[years].transpose()

df_least5.index = df_least5.index.map(int)
df_least5.plot(kind='area', alpha=0.55, figsize=(20, 10))

plt.title('Immigration Trend of Countries with Least Contribution to Immigration')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.show()
```

#### 9. Using the artist layer 

```python
df_least5 = df_usa.tail(5)

df_least5 = df_least5[years].transpose()
df_least5.head()


df_least5.index = df_least5.index.map(int)


ax = df_least5.plot(kind='area', alpha=0.65, stacked=False, figsize=(20, 10))

ax.set_title('Immigration Trend of 5 Countries with Least Contribution to Immigration')
ax.set_ylabel('Number of Immigrants')
ax.set_xlabel('Years')
```
